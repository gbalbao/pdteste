```python
#PS RACCOON utilizando pandas

from base64 import decode
from re import X
from tkinter.tix import DisplayStyle
import pandas as pd 
import requests
import urllib
import json
from IPython.display import display
import numpy as np
```

```python
#PLANILHA DIAS
url =  'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_shows'
handle = urllib.request.urlopen(url)
shows = handle.read().decode()
g = pd.read_json(shows)
g_df = pd.DataFrame(g)
display(g_df)

print(list(g.columns))
```

```python
# PLANILHA GASTOS
url2 = 'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_compras'
handle2 = urllib.request.urlopen(url2)
gastos = handle2.read().decode()
gastos_df = pd.DataFrame(data = [x.split(',') for x in gastos.split('\r\n')], columns = ['nome' , 'show', 'gastos'])
#gastos_df = pd.DataFrame(data =[x.split(',') for x in gastos.split('\r\n')] , columns = [gastos_df.iloc[0]])
#gastos_df.colums = [gastos_df.iloc[0]]
gastos_df = gastos_df.drop([0])
display(gastos_df)
#columns=None
```

```python
# PLANILHA VENDAS INGRESSOS
url3 =  'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_ingressos'
handle3 = urllib.request.urlopen(url3)
vendas = handle3.read().decode()
vendas_df = pd.DataFrame(json.loads(vendas)) #comando json.loads serve para transformar a string em lista de dicionários
display(vendas_df)
```

```python

#QUESTÃO 1 - Média de gastos com a pista
# agrupar por quem teve compra concluida e comprou pista 
# ir na planilha de gastos e somar os gastos do dia 
# para filtrar: query() (passa os filtros como parametro) ex: nome_df.query('status == concluido')
# para filtrar: loc() (passa a coluna que deseja onde esteja o filtro) ex: nome_df.loc(nomedf('status')==concluido)

# Tentativas falhas:
# display(vendas_df.loc((vendas_df['status']== 'Concluido') & (vendas_df['tipo'] == 'Pista')))
# display(vendas_df.loc((vendas_df['status'] == 'Concluido') & (vendas_df['tipo']=='Pista')))
# display(vendas_df.loc((vendas_df['status']== 'Concluido') & (vendas_df['tipo'] == 'Pista')))
# vendas_df.loc[(vendas_df["status"] == "Concluido" & vendas_df['tipo']=="Pista")]

# Forma correta
pista_compra_df = pd.DataFrame(vendas_df.loc[(vendas_df['status']=='Concluido')].loc[(vendas_df['tipo'] == 'Pista')])

#Com o dataframe acima, usar o inner para juntar essa planilha com a planilha de gastos, através do nome da pessoa (chave)
# Usando o merge

pista_gasto_df = pd.merge(pista_compra_df,gastos_df)
display(pista_gasto_df)

pista_gasto_df['gastos']= pista_gasto_df['gastos'].astype(float)
media_pista = pista_gasto_df['gastos'].mean()

print('A média de gastos com os ingressos do tipo pista foi de' , media_pista.round(2))
```

```python
# Questão 2 - Quem não compareceu ao show 
# São aqueles que tiveram as compras concluídas mas que não estão nos respectivos dias dos gastos
vc_df = vendas_df.loc[vendas_df['status']=='Concluido']

gastos_df

#pd.merge(pd.merge(vc_df,gastos_df),g_df)
pd.merge(vc_df,gastos_df)

## Substituir a coluna 'show' da planilha de gastos pelo dia - 

#gastos_renomeado = gastos_df.replace('Symbolic Toy',3)

#gastos_ren = gastos_df.replace(list(g.columns),[str(x) for x in g.loc['dia'] ])
# Comando acima deu recorreu em erro ao utilizar o merge pois a coluna 'dia' da tabela de vendas
# está em float e a de gastos estava em str

gastos_ren = gastos_df.replace(list(g.columns), list(g.loc['dia']))

gastos_ren.rename(columns ={'show':'dia'},inplace=True)
inner = pd.merge(vc_df,gastos_ren)
#display(pd.merge(vc_df,gastos_ren).head(10))

# Portanto, a tabela acima nos fornece todos aqueles que compraram com a agencia
# e foram ao show. Precisamos agora encontrar aqueles que compraram mas que não estão na tabela

outerna = pd.merge(vc_df,gastos_ren, how = 'outer')
outer = outerna.dropna(subset=['mes']) # com o dropna dessa forma conseguimos retirar as pessoas que compraram de outros concorrentes
outer['gastos'].fillna('Nenhum' , inplace = True) # substituindo valores NaN por 'Nenhum'
#display(outer)  

# o que queremos são as pessoas que tem NaN em 'gastos' mas que o status é concluido e nao nulo
display(outer.loc[outer['gastos'] == 'Nenhum']['nome']) # pessoas que compraram e não foram
```
