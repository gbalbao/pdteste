```python
#PS RACCOON utilizando pandas

from base64 import decode
from re import X
from tkinter.tix import DisplayStyle
import pandas as pd 
import requests
import urllib
import json
from IPython.display import display
import numpy as np
```

```python
#PLANILHA DIAS
url =  'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_shows'
handle = urllib.request.urlopen(url)
shows = handle.read().decode()
g = pd.read_json(shows)
g_df = pd.DataFrame(g)
display(g_df)
```

```python
# PLANILHA GASTOS
url2 = 'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_compras'
handle2 = urllib.request.urlopen(url2)
gastos = handle2.read().decode()
gastos_df = pd.DataFrame([x.split(',') for x in gastos.split('\r\n')])
display(gastos_df)
```

```python
# PLANILHA VENDAS INGRESSOS
url3 =  'https://us-central1-raccoon-bi.cloudfunctions.net/psel_de_ingressos'
handle3 = urllib.request.urlopen(url3)
vendas = handle3.read().decode()
vendas_df = pd.DataFrame(json.loads(vendas)) #comando json.loads serve para transformar a string em lista de dicionários
display(vendas_df)
```
